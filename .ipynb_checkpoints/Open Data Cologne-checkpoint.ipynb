## Offene Daten Köln
Source: https://offenedaten-koeln.de/
Themen: Lautstärke, Verkehrsdaten und Klima

```python
import json
import requests
import pandas as pd
import datetime
import numpy as np
url = "https://offenedaten-koeln.de/api/3/action/package_show?id=d367488b-fe8c-42ac-822d-113951831ead"
r = requests.get(url)
data = json.loads(r.content)
```

```python
d0 = pd.DataFrame(data['result'][0]['resources'])
df_csv = d0[d0['format'] =='csv']
df_csv = df_csv[df_csv.name.str.contains("Geschwindigkeitsueberwachung ")]
```

```python
def get_datetime(row):
    return parse_date(row["vorfallsdatum"]) + parse_time(row["vorfallsuhrzeit"])
#Datum = D/MM/YY
def parse_date(date):
    date = str(date)
    year = int("20" + date[-2:])
    month = int(date[-4:-2])
    day = int(date[:-4])
    datetime_parsed = datetime.datetime(year,month,day)
    return datetime_parsed

#Vorfallsuhrzeit ("5" = 00:05; " 335" = 03:35; "1211" = 12:11)

def parse_time(time):
    time = str(time)
    try:
        minutes = int(time[-2:])
    except:
        return np.nan
    try:
        hours = int(time[:-2])
    except:
        hours = 0
    time_parsed = datetime.timedelta(hours = hours,minutes= minutes)
    return time_parsed

def parse_hours(time):
    time = str(time)
    try:
        minutes = int(time[-2:])
    except:
        return np.nan
    try:
        hours = int(time[:-2])
    except:
        hours = 0
    return hours%24
```

```python
parse_time(5)
```

```python
df = pd.DataFrame()
for row in df_csv.url:
    monthly_data = pd.read_csv(row,delimiter=";")
    monthly_data['datetime'] = monthly_data.apply(get_datetime,axis=1)
    df = df.append(monthly_data)
```

```python
df['hours'] = df['vorfallsuhrzeit'].apply(parse_hours)
df['ueberschreitung'] = pd.to_numeric(df.ueberschreitung,errors='coerce')
```

```python
%matplotlib inline
df.plot.scatter(x='hours', y= 'ueberschreitung')
```

```python
df[df.ueberschreitung>0].groupby('hours')['ueberschreitung'].describe()
```

```python
df[df.ueberschreitung>0].head(1000).groupby('standort')['ueberschreitung'].describe().sort_values("count",ascending= False)
```
